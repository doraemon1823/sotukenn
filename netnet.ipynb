```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
G = nx.DiGraph()  # 有向グラフ (Directed Graph)
#G = nx.karate_club_graph()
df=pd.read_csv("df_DataFrame_network.csv")
```

```python
df=df.drop("Unnamed: 0",axis=1)
```

```python
list_edge=df
list_edge.loc[0][0]
```

```python
for i in range(2268):
    G.add_edge(str(list_edge.loc[i][0]),str(list_edge.loc[i][1]))
    G.add_edge(str(list_edge.loc[i][1]),str(list_edge.loc[i][0]))
nx.draw_networkx(G)
plt.show()
```
