```python
import numpy as np
import sys,os
import config2
CONSUMER_KEY    = config2.CONSUMER_KEY
CONSUMER_SECRET = config2.CONSUMER_SECRET
ACCESS_TOKEN = config2.ACCESS_TOKEN
ACCESS_TOKEN_SECRET = config2.ACCESS_TOKEN_SECRET
import tweepy
auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)
api = tweepy.API(auth,wait_on_rate_limit = True)#,wait_on_rate_limit_notify= True #PLOT TIME
#user = sys.argv[1]
import pandas as pd
```

```python
ggg=list([range(7)])*451
df2=pd.DataFrame(ggg)
o=[]
o.insert(0,"ID")
o.insert(1,"Screen_name")
o.insert(2,"Description")
o.insert(3,"Friends_count")
o.insert(4,"Follower_count")
o.insert(5,"Tweet_count")
o.insert(6,"Favourites_count")
#お気に入り登録した数
df2.columns=o
```

```python
df01=pd.read_csv("B!_id_Friends!.csv").drop("Unnamed: 0",axis=1)
df01=df01["0"]
for i in range(450):
    df2["ID"][i]=df01[i]
    go=api.get_user(df2["ID"][i])
    go0=go.screen_name
    go1=go.description
    go2=go.friends_count
    go3=go.followers_count
    go4=go.statuses_count
    go5=go.favourites_count
    df2["Screen_name"][i]=go0
    df2["Description"][i]=go1
    df2["Friends_count"][i]=go2
    df2["Follower_count"][i]=go3
    df2["Tweet_count"][i]=go4
    df2["Favourites_count"][i]=go5
```
